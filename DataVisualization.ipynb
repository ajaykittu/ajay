Pie plot

```python
import matplotlib.pyplot as plt
import numpy as np

y1 =np.array([4,9,11,29])

plt.pie(y1)
```

```python
lab=np.array(["FORD","BMW","AUDI","TATA"])
plt.pie(y1,labels=lab)
plt.show()
```

```python
plt.pie(y1,startangle=90)
plt.show()
```

```python
space=np.array([0.3,0,0,0.3])
plt.pie(y1, labels=lab,explode=space)
plt.show()
```

```python
plt.pie(y1,labels=lab,explode=space,shadow=True)
plt.show()
```

```python
plt.pie(y1,labels=lab,explode=space,shadow=True)
plt.legend()
plt.show()
```

Scatter plot

```python
import matplotlib.pyplot as plt
import numpy as np

x =np.array([4,9,11,29])
y=np.array([7,13,15,33])
plt.scatter(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x =np.array([4,9,11,29])
y=np.array([7,13,15,33])
plt.scatter(x,y)

x=np.array([9,14,19,28])
y=np.array([12,24,27,37])
plt.scatter(x,y)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x =np.array([4,9,11,29])
y=np.array([7,13,15,33])
plt.scatter(x,y,color='red')

x=np.array([9,14,19,28])
y=np.array([12,24,27,37])
plt.scatter(x,y,color="green")

plt.show()
```

```python
colors=np.array(["VIOLET","GREY","YELLOW","BROWN"])
plt.scatter(x,y,c=colors)
plt.show()
```

```python
plt.scatter(x,y,c=colors,cmap='virids')
plt.colorbar()
plt.show()
```

```python
sizes=np.array([250,500,750,1000])
plt.scatter(x,y, s=sizes,alpha=0.3)
plt.show()
```

```python
x=np.random.randint(100, size=(100))
y=np.random.randint(100, size=(100))
colors=np.random.randint(100,size=(100))
sizes=10 * np.random.randint(100,size=(100))
plt.scatter(x,y,c=colors,s=sizes,alpha=0.3,cmap='nipy_spectral')
plt.colorbar()
plt.show()
```

Bar plot

```python
import matplotlib.pyplot as plt
import numpy as np

x =np.array(["FORD","BMW","AUDI","TATA"])
y =np.array([5,10,15,20])

plt.bar(x,y)
plt.show()
```

```python
plt.bar(x,y,width=0.5)
```

```python
plt.bar(x,y,width=2)
plt.show()
```

```python
plt.bar(x,y,width=0.5,color="red")
plt.show()
```

```python
plt.barh(x,y,color="orange",height=0.5)
plt.show()
```

```python
plt.barh(x,y,color="orange",height=2)
plt.show()
```

Histogram

```python
import matplotlib.pyplot as plt
plt.title('Students Performance')
plt.xlabel('Marks Groups')
plt.ylabel('No. of Students')
marks=[22,16,23,49,39,32,48,37,29,28,21,37,26,19,18,38,45,19,49,25]
plt.hist(marks,edgecolor='black')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.title('Students Performance')
plt.xlabel('Marks Groups')
plt.ylabel('No. of Students')
marks=[22,16,23,49,39,32,48,37,29,28,21,37,26,19,18,38,45,19,49,25]
plt.hist(marks,edgecolor='black',bins=5)
plt.show()
```

```python
plt.hist(marks,edgecolor='black',bins=5,color='red')
plt.show()
```

```python
bins=[0,10,20,30,40,50]
plt.hist(marks,edgecolor='black',bins=bins,color='red')
plt.show()
```
